# IOCaml support for numcaml

```ocaml
#require "numcaml.syntax";;
#require "iocaml-kernel.notebook";;
```

Setup numcaml for the notebook (this could be done with a .iocamlinit file).  Open the library and install a printer for `Numcaml.Math.t`.

```ocaml
open Numcaml
```

```ocaml
let numcaml_printer fmt x = 
    Iocaml.display "text/html" (Math.mathjax x);
    Format.fprintf fmt "<mathjax>"
```

```ocaml
#install_printer numcaml_printer
```
